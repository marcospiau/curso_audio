# Erro de Quantização

```python
# Erro de quantizacao

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

bits = 3
vmin = 0
vmax = 10

divisoes = 2**bits

vin = np.linspace(vmin-1.0, vmax+1.0, 1000)
x = np.floor(((vin - vmin) / (vmax-vmin)) * divisoes)
x = np.minimum(x, divisoes-1)
x = np.maximum(x, 0)
plt.plot(vin, x)
plt.xlabel('Tensao de entrada (V)')
plt.ylabel('Representacao digital')
plt.show()
```

```python
bits = 3
vmin = 0
vmax = 10

divisoes = 2**bits

t = np.linspace(0, 10, 1000)
vin = (np.cos(2 * np.pi * 0.2 * t) * 7) + 5


x = np.floor(((vin - vmin) / (vmax-vmin)) * divisoes)
x = np.minimum(x, divisoes-1)
x = np.maximum(x, 0)
plt.plot(t, vin, ':', label='Sinal original')
plt.plot(t, x, label='Sinal amostrado')
plt.legend()
plt.xlabel('Tempo (s)')
plt.ylabel('Magnitude')
plt.show()
```
